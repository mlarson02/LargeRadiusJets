```python
import matplotlib.pyplot as plt
from adjustText import adjust_text

def customizable_two_line_graph(
    x1, y1, captions1, label1, color1,
    x2, y2, captions2, label2, color2,
    title='Line Graph',
    xlabel='X-axis',
    ylabel='Y-axis',
    marker1='o',
    marker2='s'
):
    plt.figure(figsize=(10, 6))
    plt.plot(x1, y1, color=color1, marker=marker1, label=label1)
    plt.plot(x2, y2, color=color2, marker=marker2, label=label2)
    texts = []
    for xi, yi, caption in zip(x1, y1, captions1):
        texts.append(plt.text(xi, yi, caption, fontsize=9, color=color1))
    for xi, yi, caption in zip(x2, y2, captions2):
        texts.append(plt.text(xi, yi, caption, fontsize=9, color=color2))
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    # Adjust all text labels to avoid overlap
    adjust_text(texts, arrowprops=dict(arrowstyle='->', color='red'))
    plt.show()

# Example usage
x1 = [1, 2, 4, 16, 32, 64, 128]
y1 = [312, 184, 173, 184, 184, 184, 90]
captions1 = ['II = 1', 'II = 1', 'II = 2', 'II = 8', 'II = 16', 'II = 32', '']

x2 = [1, 2, 4, 8, 32, 64, 128]
y2 = [312, 184, 184, 173, 184, 184, 89]
captions2 = ['II = 1', 'II = 1', 'II = 2', 'II = 4', 'II = 16', 'II = 32', '']

customizable_two_line_graph(
    x1, y1, captions1, 'Partition Factor 4', 'green',
    x2, y2, captions2, 'Partition Factor 8', 'blue',
    title='Unroll Factor vs. Latency',
    xlabel='Unroll Factor',
    ylabel='Latency (clock cyles)'
)
```

```python
def customizable_one_line_graph(
    x, y, captions, color,
    title='Line Graph',
    xlabel='X-axis',
    ylabel='Y-axis',
    marker='o'
):
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, color=color, marker=marker)
    for xi, yi, caption in zip(x, y, captions):
        plt.text(xi, yi, caption, fontsize=9, color=color)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

x1 = [1, 2, 4, 8, 16, 32, 64, 128]
y1 = [130, 66, 34, 18, 10, 6, 4, 1]
captions1 = []


customizable_one_line_graph(
    x1, y1, captions1, 'blue',
    title='Bitmask loop2 Unroll Factor vs. Latency',
    xlabel='Unroll Factor',
    ylabel='Latency (clock cyles)'
)
```
